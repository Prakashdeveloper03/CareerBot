```python
import pandas as pd

df = pd.read_csv("../data/keywords.csv")
df.head()
```

```python
# Exclude rows where the 'Skill' column values are 'r', 'c', 'cc', or 'go'
df = df[~df["Skill"].isin(["r", "c", "cc", "go"])]
df.head()
```

```python
# Group by job title and aggregate skills into a single column separated by commas
grouped_df = (
    df.groupby("Job Title")["Skill"].apply(lambda x: ", ".join(x)).reset_index()
)
grouped_df.rename(columns={"Job Title": "title", "Skill": "skills"}, inplace=True)
grouped_df.head()
```

```python
# Save the grouped_df inplace of original csv file
grouped_df.to_csv("../data/keywords.csv", index=False)
```
